```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
col_names=['sno','Gender','Age','Symptom1','Symptom2','Symptom3','Symptom4','Symptom5','Symptom6','Symptom7','Disease']
df=pd.read_csv('patient.csv',names=col_names)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.Disease=le.fit_transform(df.Disease)
df.head()
```

```python
x=df.Age
y=df.Disease
plt.pie(y,labels=x)
```
